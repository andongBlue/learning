```python
bins=[0,100,300,600,10000]
levels1=['0-100','100-300','300-600','600-10000']
data['company_size']=pd.cut(data['18年人员'],bins,labels=levels1)
data.head()
```

```python
data['colors']=['red' if x<25 else 'green' for x in data.R]
plt.figure(figsize=(15,6),dpi=100)
#可以将xmin设为非0得一个数值,即分界线
plt.hlines(y=data.厂家,xmin=25,xmax=data.R,color=data.colors,alpha=0.6,linewidth=6)

#将图表美化
plt.title('供方生命周期', fontdict={'size':20})
#plt.gcf()和plt.gca()获得，分别表示Get Current Figure和Get Current Axes。在pyplot模块中
#许多函数都是对当前的Figure或Axes对象进行处理
#比如说：plt.plot()实际上会通过plt.gca()获得当前的Axes对象ax，然后再调用ax.plot()方法实现真正的绘图
plt.grid(linestyle='--',alpha=0.4)

#添加文本
for x, y, tex in zip(data.R, data.index, data.R):
    t = plt.text(x, y, round(tex, 2), horizontalalignment='right' if x < 25 else 'left',verticalalignment='center', fontdict={'color':'black' if x < 25
else 'green', 'size':10})
```
