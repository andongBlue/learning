```python
#为了保持数据的统一，将数据列标准化
from sklearn.preprocessing import StandardScaler
x_train=StandardScaler().fit_transform(x_train)
print(x_train.shape)
print(len(y_train))
x_train
```

```python
#将训练数据拆分为训练集和测试集,KFold,所有的数据用来训练模型
from sklearn.model_selection import KFold
def get_predict(x,y,clf_class,**kwargs):
    y_pred=y.copy()
    kf=KFold(n_splits=10,shuffle=True,random_state=42)

    for train_index,test_index in kf.split(x):
        x_train=x[train_index]
        y_train=y[train_index]
        x_test=x[test_index]
        clf=clf_class(**kwargs)
        clf.fit(x_train,y_train)
        y_pred[test_index]=clf.predict(x_test)
        
    return y_pred
```

```python
#定义一个取得预测概率的函数
def get_predict_proba(x,y,clf_class,**kwargs):
    p=np.zeros((len(y),2))
    kf=KFold(n_splits=10,shuffle=True,random_state=42)

    for train_index,test_index in kf.split(x):
        x_train=x[train_index]
        y_train=y[train_index]
        x_test=x[test_index]
        clf=clf_class(**kwargs)
        clf.fit(x_train,y_train)
        p[test_index]=clf.predict_proba(x_test)
        
    return p
```

```python
#引入 SVM和KNN进行分类训练
from sklearn.svm import SVC as svm
from sklearn.neighbors import KNeighborsClassifier as knn
```

```python
pred_svm=get_predict(x_train,y_train,svm,probability=True)
pred_knn=get_predict(x_train,y_train,knn)
```

```python
#引入查准率precision score和查全率recall score对两种算法训练的效果进行测算
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import roc_auc_score
```

```python
#测试两种模型的效果 精度
print('accuracy score of SVM is:%.3f'%accuracy_score(y_train,get_predict(x_train,y_train,svm)))  
print('accuracy score of knn is:%.3f'%accuracy_score(y_train,get_predict(x_train,y_train,knn)))  
```

```python
#测试两种模型的效果 查全率
print('recall score of SVM is:%.3f'%recall_score(y_train,get_predict(x_train,y_train,svm)))  
print('recall score of knn is:%.3f'%recall_score(y_train,get_predict(x_train,y_train,knn)))  
```

```python
#测试两种模型的效果 查准率
print('precision score of SVM is:%.3f'%precision_score(y_train,get_predict(x_train,y_train,svm)))  
print('precision score of knn is:%.3f'%precision_score(y_train,get_predict(x_train,y_train,knn)))  
```

```python
#测试两种模型的效果 ROC-面积
print('ROC score of SVM is:%.3f'%roc_auc_score(y_train,get_predict_proba(x_train,y_train,svm,probability=True)[:,1]))  
print('ROC score of knn is:%.3f'%roc_auc_score(y_train,get_predict_proba(x_train,y_train,knn)[:,1]))  
```

以上分数均很高,模型训练的效果不错,进一步将分类效果可视化

```python
#引入混淆矩阵对分类的结果进行可视化
from sklearn.metrics import confusion_matrix
import itertools
def plot_confusion_matrix(cm, classes,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=0)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
#compute 混淆矩阵
cnf_with_svm=confusion_matrix(y_train,get_predict(x_train,y_train,svm))
cnf_with_knn=confusion_matrix(y_train,get_predict(x_train,y_train,knn))
np.set_printoptions(precision=2)
classes_names=[0,1]
```

```python
#分类器的混淆矩阵进行可视化，可以看到分类的效果比好
plt.figure(figsize=(14,4))

plt.subplot(121)
plot_confusion_matrix(cnf_with_svm,classes_names)
plt.title('svm')

plt.subplot(122)
plot_confusion_matrix(cnf_with_knn,classes_names)
plt.title('knn')
```

```python
#定义函数绘制ROC曲线
from sklearn.metrics import roc_curve
from sklearn import metrics
def auc_curve(y,prob):
    fpr,tpr,threshold = roc_curve(y,prob) ###计算真正率和假正率
    roc_auc = metrics.auc(fpr,tpr) ###计算auc的值 
    plt.figure()
    lw = 2
    plt.figure(figsize=(7,5))
    plt.plot(fpr, tpr, color='darkorange',
             lw=lw, label='ROC curve (area = %0.3f)' % roc_auc) ###假正率为横坐标，真正率为纵坐标做曲线
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC curve')
    plt.legend(loc="lower right")
 
    plt.show()
```
