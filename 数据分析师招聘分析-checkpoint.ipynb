```python
def cut_word(word):
    position=word.find('-')
    length=len(word)
    if position!=-1:
        bottomsalary=word[:position-1]
    else:
        bottomsalary=word[:word.upper().find('K')]
    return bottomsalary
```

```python
df['bottomsalary']=df.salary.apply(cut_word)
```

```python
df['bottomsalary']=df.bottomsalary.astype('float')
```

```python
def cut_word(word,method):
    position=word.find('-')
    length=len(word)
    if position!=-1:
        bottomsalary=word[:position-1]
        topsalary=word[position+1:length-1]
    else:
        bottomsalary=word[:word.upper().find('K')]
        topsalary=bottomsalary
    if method=='bottom':
        return bottomsalary
    else:
        return topsalary
```

```python
df['topsalary']=df.salary.apply(cut_word,method='top')
```

```python
df['topsalary']=df.topsalary.astype('float')
```

```python
df.head()
```

```python
df['avesalary']=df.apply(lambda x:(x.bottomsalary+x.topsalary)/2,axis=1)
```
